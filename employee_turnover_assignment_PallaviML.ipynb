```python
#3.1.	Choose columns satisfaction_level, last_evaluation, and left.
# Select the columns 'satisfaction_level', 'last_evaluation', and 'left'
selected_columns = data[['satisfaction_level', 'last_evaluation', 'left']]

# Filter the data to include only employees who left
left_employees = selected_columns[selected_columns['left'] == 1]
kfeatures = left_employees
```

```python
# 1. Identify how many number of hidden categories are present.

# Get WCSS (WithIn Cluster Sum of Square) -- Error Metric

# THE ELBOW METHOD
from sklearn.cluster import KMeans

wcss = []

for cluster in range(2,10):
  model = KMeans(n_clusters=cluster)
  model.fit(kfeatures)
  wcss.append(model.inertia_)

plt.plot(range(2,10) , wcss)
```

```python

#Model Building

# Perform K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
left_employees['cluster'] = kmeans.fit_predict(left_employees[['satisfaction_level', 'last_evaluation']])
```

```python
# Plot the clusters
plt.figure(figsize=(10, 6))
plt.scatter(left_employees['satisfaction_level'], left_employees['last_evaluation'], c=left_employees['cluster'], cmap='viridis')
plt.xlabel('Satisfaction Level')
plt.ylabel('Last Evaluation')
plt.title('Clustering of Employees Who Left Based on Satisfaction and Evaluation')
plt.show()

# Display the first few rows of the clustered data
print("Clustered Data:****************************")
print(left_employees.head())
```

```python
#5.1.	Train a logistic regression model, apply a k-fold CV, and plot the classification report.
#get score using logistic regression model
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import classification_report

model = LogisticRegression(max_iter=1000 , random_state=123)

scores = cross_val_score(model,
                         resampledFeatures,
                         resampledLabel,
                         cv=5)

print("Logisitcs Regression Model Score : " , scores)
print("Mean Score : ", scores.mean())

predicted_labels = cross_val_predict(model, resampledFeatures, resampledLabel, cv=5)
# Generate the classification report
class_report = classification_report(resampledLabel, predicted_labels, output_dict=True)

# Convert the classification report to a DataFrame for better visualization
class_report_df = pd.DataFrame(class_report).transpose()

# Plot the classification report
plt.figure(figsize=(10, 6))
sns.heatmap(class_report_df.iloc[:-1, :-1], annot=True, cmap='Blues')
plt.title('Classification Report for Logistic Regression (k-Fold CV)')
plt.show()

```

```python
#6.1.	Find the ROC/AUC for each model and plot the ROC curve.
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Load the CSV file
curvDF = pd.read_csv('HR_comma_sep.csv')

# Pre-process the data by converting categorical columns to numerical columns
curvDF = pd.get_dummies(curvDF, columns=['sales', 'salary'], drop_first=True)

# Separate features and target variable
X = curvDF.drop('left', axis=1)
y = curvDF['left']

# Split the data into train and test sets (80:20 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Initialize models
logistic_regression = LogisticRegression(max_iter=1000)
random_forest = RandomForestClassifier(random_state=123)
gradient_boosting = GradientBoostingClassifier(random_state=123)

# Train the models
logistic_regression.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
gradient_boosting.fit(X_train, y_train)

# Predict probabilities for the test set
y_pred_proba_lr = logistic_regression.predict_proba(X_test)[:, 1]
y_pred_proba_rf = random_forest.predict_proba(X_test)[:, 1]
y_pred_proba_gb = gradient_boosting.predict_proba(X_test)[:, 1]

# Calculate ROC curves and AUC scores
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_pred_proba_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)

fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_proba_rf)
roc_auc_rf = auc(fpr_rf, tpr_rf)

fpr_gb, tpr_gb, _ = roc_curve(y_test, y_pred_proba_gb)
roc_auc_gb = auc(fpr_gb, tpr_gb)



# Plot ROC curves
plt.figure(figsize=(10, 6))
plt.plot(fpr_lr, tpr_lr, color='blue', lw=2, label=f'Logistic Regression (AUC = {roc_auc_lr:.2f})')
plt.plot(fpr_rf, tpr_rf, color='green', lw=2, label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot(fpr_gb, tpr_gb, color='red', lw=2, label=f'Gradient Boosting (AUC = {roc_auc_gb:.2f})')
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

# Print AUC scores for each model
print(f"Logistic Regression AUC: {roc_auc_lr:.2f}")
print(f"Random Forest AUC: {roc_auc_rf:.2f}")
print(f"Gradient Boosting AUC: {roc_auc_gb:.2f}")


# Print the observations
print("Observations: \n")
print("I calculated the ROC/AUC for each model and plotted the ROC curve.")
print("Here are the results:\n")
print("ROC AUC Scores")
print(f"Logistic Regression AUC: {roc_auc_lr:.2f}")
print(f"Random Forest AUC: {roc_auc_rf:.2f}")
print(f"Gradient Boosting AUC: {roc_auc_gb:.2f}")
```

```python
# 6.2.Find the confusion matrix for each of the models.
#find the confusion matrix for each of the models (Logistic Regression, Random Forest, and Gradient Boosting)

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import confusion_matrix

# Load the CSV file
df = pd.read_csv('HR_comma_sep.csv')

# Pre-process the data by converting categorical columns to numerical columns
df = pd.get_dummies(df, columns=['sales', 'salary'], drop_first=True)

# Separate features and target variable
X = df.drop('left', axis=1)
y = df['left']

# Split the data into train and test sets (80:20 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Initialize models
logistic_regression = LogisticRegression(max_iter=1000)
random_forest = RandomForestClassifier(random_state=123)
gradient_boosting = GradientBoostingClassifier(random_state=123)

# Train the models
logistic_regression.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
gradient_boosting.fit(X_train, y_train)

# Predict on the test set
y_pred_lr = logistic_regression.predict(X_test)
y_pred_rf = random_forest.predict(X_test)
y_pred_gb = gradient_boosting.predict(X_test)

# Calculate confusion matrices
confusion_matrix_lr = confusion_matrix(y_test, y_pred_lr)
confusion_matrix_rf = confusion_matrix(y_test, y_pred_rf)
confusion_matrix_gb = confusion_matrix(y_test, y_pred_gb)

# Print confusion matrices for each model
print("Confusion Matrix for Logistic Regression:")
print(confusion_matrix_lr)
print("\nConfusion Matrix for Random Forest:")
print(confusion_matrix_rf)
print("\nConfusion Matrix for Gradient Boosting:")
print(confusion_matrix_gb)

print("Observations:\n")

#Logistic Regression
print("Logistic Regression:")
print("True Negatives (TN): 2135")
print("False Positives (FP): 156")
print("False Negatives (FN): 475")
print("True Positives (TP): 234\n")

#Random Forest
print("Random Forest:")
print("True Negatives (TN): 2286")
print("False Positives (FP): 5")
print("False Negatives (FN): 18")
print("True Positives (TP): 691\n")

#Gradient Boosting
print("Gradient Boosting:")
print("True Negatives (TN): 2265")
print("False Positives (FP): 26")
print("False Negatives (FN): 48")
print("True Positives (TP): 661\n")

#Conclusion
print("The Random Forest model has the best performance with the highest number of true positives and true negatives,")
print("and the lowest number of false positives and false negatives. This aligns with the ROC AUC scores,")
print("confirming that Random Forest is the best model for predicting employee turnover.")
```

```python
#7.2.	Based on the probability score range below, categorize the employees into four zones and suggest your thoughts on the retention strategies for each zone.
#■	Safe Zone (Green) (Score < 20%)
#■	Low-Risk Zone (Yellow) (20% < Score < 60%)
#■	Medium-Risk Zone (Orange) (60% < Score < 90%)
#■	High-Risk Zone (Red) (Score > 90%).

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Load the CSV file
df = pd.read_csv('HR_comma_sep.csv')

# Pre-process the data by converting categorical columns to numerical columns
df = pd.get_dummies(df, columns=['sales', 'salary'], drop_first=True)

# Separate features and target variable
X = df.drop('left', axis=1)
y = df['left']

# Split the data into train and test sets (80:20 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# Initialize the best model (Random Forest)
random_forest = RandomForestClassifier(random_state=123)
random_forest.fit(X_train, y_train)

# Predict probabilities for the test set
y_pred_proba = random_forest.predict_proba(X_test)[:, 1]

# Categorize employees into risk zones based on probability scores
risk_zones = pd.cut(y_pred_proba, bins=[0, 0.2, 0.6, 0.9, 1], labels=['Safe Zone', 'Low-Risk Zone', 'Medium-Risk Zone', 'High-Risk Zone'])

# Add risk zones to the test set
X_test['Risk Zone'] = risk_zones

# Display the first few rows of the test set with risk zones
print(X_test.head())

# Count the number of employees in each risk zone
risk_zone_counts = X_test['Risk Zone'].value_counts()
print("\nNumber of employees in each risk zone:")
print(risk_zone_counts)

# Suggest retention strategies for each zone
retention_strategies = {
    'Safe Zone': [
        "Recognition and Rewards: Continue recognizing and rewarding their contributions to maintain high morale.",
        "Career Development: Offer opportunities for career growth and development to keep them motivated.",
        "Work-Life Balance: Ensure a healthy work-life balance to maintain their satisfaction levels."
    ],
    'Low-Risk Zone': [
        "Regular Feedback: Conduct regular feedback sessions to address any concerns and provide support.",
        "Professional Development: Offer training and development programs to enhance their skills and career prospects.",
        "Engagement Activities: Organize team-building and engagement activities to strengthen their connection with the company."
    ],
    'Medium-Risk Zone': [
        "One-on-One Meetings: Schedule one-on-one meetings to understand their concerns and provide personalized support.",
        "Mentorship Programs: Implement mentorship programs to guide and support their career development.",
        "Work Environment: Improve the work environment by addressing any issues related to management, workload, or team dynamics."
    ],
    'High-Risk Zone': [
        "Immediate Intervention: Take immediate action to address their concerns and provide solutions.",
        "Retention Bonuses: Offer retention bonuses or incentives to encourage them to stay.",
        "Career Pathing: Provide clear career paths and opportunities for advancement to retain their interest and commitment.",
        "Flexible Work Options: Offer flexible work arrangements, such as remote work or flexible hours, to improve their work-life balance."
    ]
}

print("\nRetention Strategies for Each Zone:")
for zone, strategies in retention_strategies.items():
    print(f"\n{zone}:")
    for strategy in strategies:
        print(f"- {strategy}")


```
